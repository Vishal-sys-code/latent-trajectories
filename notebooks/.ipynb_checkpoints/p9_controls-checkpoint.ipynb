# Phase 4: Control Experiments

This notebook computes the control experiments (Random Embeddings, Layer Shuffle) to ensure that the observed trajectory structure is not an artifact.

```python
import pandas as pd
import numpy as np
import os
from sklearn.metrics import silhouette_score

# Ensure results directory exists
os.makedirs('../results/controls', exist_ok=True)

def compute_controls():
    df_pca = pd.read_parquet('../results/projected/gpt2_pca.parquet')
    layers = sorted(df_pca['layer'].unique())
    
    real_silhouette = []
    random_embeddings_silhouette = []
    layer_shuffle_silhouette = []
    
    # Baseline structure
    for l in layers:
        df_l = df_pca[df_pca['layer'] == l]
        if len(df_l['group'].unique()) > 1:
            X = df_l[['x', 'y', 'z']].values
            labels = df_l['group'].values
            
            # Real score
            score = silhouette_score(X, labels)
            real_silhouette.append({'layer': l, 'score': score})
            
            # Control A: Random Embeddings (simulate with random normal)
            X_rand = np.random.normal(0, np.std(X), X.shape)
            rand_score = silhouette_score(X_rand, labels)
            random_embeddings_silhouette.append({'layer': l, 'score': rand_score})
    
    # Control C: Layer Shuffle
    # Shuffle the representation mapping to layers across the dataset
    df_pca_shuffled = df_pca.copy()
    np.random.seed(42)
    df_pca_shuffled['layer'] = np.random.permutation(df_pca['layer'].values)
    
    for l in layers:
        df_l_shuffled = df_pca_shuffled[df_pca_shuffled['layer'] == l]
        if len(df_l_shuffled['group'].unique()) > 1:
            X_shuffled = df_l_shuffled[['x', 'y', 'z']].values
            labels_shuffled = df_l_shuffled['group'].values
            if len(np.unique(labels_shuffled)) > 1:
                shuffled_score = silhouette_score(X_shuffled, labels_shuffled)
                layer_shuffle_silhouette.append({'layer': l, 'score': shuffled_score})
        
    df_real = pd.DataFrame(real_silhouette)
    df_rand = pd.DataFrame(random_embeddings_silhouette)
    df_shuffled_layer = pd.DataFrame(layer_shuffle_silhouette)
    
    df_real.to_csv('../results/controls/gpt2_real_silhouette.csv', index=False)
    df_rand.to_csv('../results/controls/gpt2_random_embeddings.csv', index=False)
    df_shuffled_layer.to_csv('../results/controls/gpt2_layer_shuffle.csv', index=False)
    
    print("Controls computed and saved.")

if __name__ == '__main__':
    compute_controls()
```
